## **2.4 ARRAY (Vetor e Matriz)**üî¢
---

&emsp;Um **array** (ou vetor \ matriz) √© uma estrutura de dados **linear** e **est√°tica**que armazena elementos do **mesmo tipo** em posi√ß√µes consecutivas da mem√≥ria. √ötil quando precisamos armazenar e acessar m√∫ltiplos valores rapidamente.

>üí≠&emsp; **Exemplo:**

>üìå**Caracter√≠sticas do Array**
--
- ‚úÖHomogeneidade: Todos os elementos do array s√£o do mesmo tipo.

- ‚úÖAcesso R√°pido: O acesso a um elemento √© feito diretamente pelo √≠ndice, o que o torna muito eficiente.

- ‚úÖTamanho Fixo: Em algumas linguagens (como C e Java), o tamanho do array √© fixo e definido na cria√ß√£o. Em Python, as listas s√£o flex√≠veis e podem crescer dinamicamente.

>üìå**Conceitos Fundamentais:**
--

- **Elementos:** Cada valor armazenado no array √© chamado de elemento.

- **√çndices:** Cada elemento do array possui um √≠ndice, que √© um n√∫mero inteiro que representa sua posi√ß√£o na lista. Na maioria das linguagens de programa√ß√£o, os √≠ndices come√ßam em 0.

- **Tipo de Dados:** Em geral, os arrays armazenam elementos do mesmo tipo de dado (inteiros, strings, etc.).

- **Tamanho Fixo:** Em arrays est√°ticos, o tamanho √© definido na cria√ß√£o e n√£o pode ser alterado. J√° os arrays din√¢micos podem crescer ou diminuir conforme necess√°rio.

>üìå **Tipos de Arrays**
--
1. **Array Unidimensional** (Vetor): Uma lista de elementos em uma √∫nica linha.
<br>
2. **Array Multidimensional** (Matriz): Uma tabela com v√°rias linhas e colunas.

---

**‚û°Ô∏è Exemplo de Array em Python**
--

**1Ô∏è‚É£ Array Unidimensional (Vetor)**
--

>**Obs**: Em Python, usamos listas para representar arrays,
pois elas s√£o mais flex√≠veis.

```python
# Criando um array (lista) de n√∫meros
numeros = [10, 20, 30, 40, 50]

# Acessando elementos pelo √≠ndice
print(numeros[0])  # Sa√≠da: 10
print(numeros[2])  # Sa√≠da: 30

# Modificando um elemento
numeros[1] = 25
print(numeros)  # Sa√≠da: [10, 25, 30, 40, 50]
```

**2Ô∏è‚É£ Array Bidimensional (Matriz)**
--

>**Dica:** Matrizes s√£o √∫teis para representar tabuleiros de jogos, imagens e dados organizados em tabelas.

```python
# Criando uma matriz 3x3 (3 linhas e 3 colunas)
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Acessando um elemento (linha 1, coluna 2)
print(matriz[1][2])  # Sa√≠da: 6

# Modificando um elemento
matriz[0][1] = 99
print(matriz)
```

‚û°Ô∏è **Opera√ß√µes B√°sicas**
--
>Usando o M√≥dulo array (Para tipos espec√≠ficos de dados)

```python
import array

"""
1Ô∏è‚É£ Criando um Array
"""
# 'i' indica n√∫meros inteiros
arr = array.array('i', [10, 20, 30, 40, 50])
print('Criando um Array:\n',arr)  # Sa√≠da: array('i', [10, 20, 30, 40, 50])

"""
2Ô∏è‚É£Acessando Elementos
"""
print('Acessando Primeiro elemento:\n',arr[0])  # Primeiro elemento ‚Üí Sa√≠da: 10
print('Acessando Terceiro elemento:\n',arr[2])  # Terceiro elemento ‚Üí Sa√≠da: 30
print('Acessando Terceiro elemento:\n',arr[-1]) # √öltimo elemento ‚Üí Sa√≠da: 50


"""
3Ô∏è‚É£Modificando Elementos
"""
arr[1] = 25
print('\nModificando Elementos:\n',arr)  # Sa√≠da: array('i', [10, 25, 30, 40, 50])


"""
4Ô∏è‚É£Adicionando Elementos
"""
# append() - Adiciona no final
arr.append(60)
print('\nAdicionando Elementos (append())\n',arr)  # Sa√≠da: [10, 25, 99, 40, 50, 60]


# insert() - Insere em uma posi√ß√£o espec√≠fica
arr.insert(3, 99)
print('\nAdicionando Elementos (insert())\n',arr)  # Sa√≠da: [10, 25, 99, 40, 99, 50, 60]


"""
5Ô∏è‚É£Removendo Elementos
"""
# remove() - Remove um valor espec√≠fico
arr.remove(99)
print('\nRemovendo Elementos (remove(99)):\n',arr)  # Sa√≠da: [10, 25, 40, 99, 50, 60]

# pop() - Remove e retorna o √∫ltimo elemento (ou o de um √≠ndice espec√≠fico)
ultimo = arr.pop()
print('\nRemovendo Elementos pop()\n',ultimo)  # Sa√≠da: 60
print(arr)  # Sa√≠da: [10, 25, 40, 99, 50]


"""
6Ô∏è‚É£Tamanho do Array
"""
print('\nTamanho do Array:',len(arr))  # Sa√≠da: 5


"""
7Ô∏è‚É£Percorrendo o Array
"""
for elemento in arr:
    print('\nPercorrendo o Array:',elemento)


"""
8Ô∏è‚É£Concatenando Arrays
"""
arr2 = array.array('i', [70, 80, 90])
arr3 = arr + arr2
print(arr3)


"""
9Ô∏è‚É£Verificando a Exist√™ncia de um Elemento
"""
print('\nVerificando a Exist√™ncia de um Elemento')
if 40 in arr:
    print("40 est√° no array")
```
